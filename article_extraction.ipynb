```python
import os
from goose3 import Goose
from goose3.network import NetworkError
import pandas as pd

os.makedirs('data_extraction_new', exist_ok=True)
def extract_article_and_title(url):
    """Extracts the article text and title from a given URL using Goose3.

    Args:
        url (str): The URL of the web page to extract data from.

    Returns:
        tuple: A tuple containing the article text and title.
    """
    g=Goose()
    try:
        article = g.extract(url=url)
        return article.cleaned_text, article.title
    except NetworkError as e:
        print(f"NetworkError occurred: {e}")
        return None, None

# Example usage
#url = "https://insights.blackcoffer.com/ml-and-ai-based-insurance-premium-model-to-predict-premium-to-be-charged-by-the-insurance-company/"
#df=pd.read_csv("/content/Input.csv")
#for i in df.URL:
with open('/content/Input.csv', 'r', encoding='utf-8') as csvfile:
    next(csvfile)  # Skip the header row
    for line in csvfile:
        url_id, url = line.strip().split(',')
        article_text, article_title = extract_article_and_title(url)
  
        with open(f'data_extraction_new/{url_id}.txt', 'w', encoding='utf-8') as outfile:
          outfile.write(article_title + '\n\n')
          outfile.write(article_text)
        print(f'Article saved to {url_id}.txt')





```
